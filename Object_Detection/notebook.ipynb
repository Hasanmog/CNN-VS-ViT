```python
import torch


bbox = [100 , 200 , 300 , 400]
x0, y0, x1, y1 = bbox
cat = 2
z = torch.zeros((6 , 512 , 512))
z[cat, bbox[0]:bbox[2], bbox[1]:bbox[3]] = 1  # Note the correct order for slicing

x0
```

```python
import matplotlib.pyplot as plt

# Visualize the tensor slice for the specified category
plt.imshow(z[cat], cmap='gray')
plt.title(f'Category {cat} - Bounding Box Region')
plt.colorbar()
plt.show()
```
