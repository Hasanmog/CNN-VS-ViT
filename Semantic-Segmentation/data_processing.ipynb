```python
import os
import random
```

```python
dataset_dir = "../Datasets/Underwater"
images = os.path.join(dataset_dir , "train_val/images")
masks = os.path.join(dataset_dir , "train_val/masks" )
```

```python
print(f"Total number of images = {len(os.listdir(images))}")
print(f"Total number of masks = {len(os.listdir(masks))}")
```

```python
images = os.listdir(images)
masks = os.listdir(masks)

random.shuffle(images)

val = images[1300:]
```

```python
names = [name.replace(".jpg" , '') for name in images]

val_masks = []

# Iterate through each mask and check if it starts with any prefix in 'names'
for mask in masks:
    if any(mask.startswith(name) for name in names):
        val_masks.append(mask)
```
